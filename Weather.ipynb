```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("weather.csv")
df = pd.DataFrame(data)
```

```python
data.head()
```

```python
data.info()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df["Wind Speed (km/h)"], labels=df["Country"], autopct='%1.1f%%', startangle=140)
plt.title("Wind Speed by Country")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.fill_between(df["Country"], df["Temperature"], color='skyblue')
plt.title("Temperature by Country")
plt.xlabel("Country")
plt.ylabel("Temperature (C)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df["Country"], df["Precipitation (mm)"], color='lightcoral')
plt.title("Precipitation by Country")
plt.xlabel("Country")
plt.ylabel("Precipitation (mm)")
plt.xticks(rotation=45)
plt.show()
```
